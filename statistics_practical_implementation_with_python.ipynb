```python
def variance(data):
    n=len(ages)
    # mean of the data
    mean=sum(data)/n
    # variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-1)
    return variance
```

```python
variance(ages)
```

```python
def variance(data):   # population data
    n=len(ages)
    # mean of the data
    mean=sum(data)/n
    # variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/n
    return variance
```

```python
variance(ages)
```

```python
def variance(data,dof=0):   # population data
    n=len(ages)
    # mean of the data
    mean=sum(data)/n
    # variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-dof)
    return variance
```

```python
variance(ages)
```

```python
def variance(data,dof=1):   # sample data
    n=len(ages)
    # mean of the data
    mean=sum(data)/n
    # variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-dof)
    return variance
```

```python
variance(ages)
```

###  z score

```python
outliers=[]
def detect_outliers(data):
    threshold=3  # third standard deviation
    mean=np.mean(data)
    std=np.std(data)
    
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
detect_outliers(dataset)
```
